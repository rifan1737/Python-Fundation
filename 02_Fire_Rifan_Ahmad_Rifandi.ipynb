```python
def calculate_sugih():
    capital = int(input('Input money capital: '))
    tenure = int(input('Input investment tenure (year): '))

    # First and second conditions for maximum money capital and tenure limit
    if capital < 0 or capital > 1000000000:
        print('Invalid capital amount. Please enter a capital between 0 and 1,000,000,000.')
    elif tenure < 1 or tenure > 25:
        print('Invalid tenure. Please enter a tenure between 1 and 25 years.')
    else:
        year = 1
        total = capital

        # Begin the iteration using a while loop
        while year <= tenure:
            total += total * 0.25  # Calculate the cumulative investment
            print('Year {}, your total investment: {}'.format(year, total))
            year += 1

# Don't change anything here
calculate_sugih()


```
